## 1.1 Basic Bar Chart:
to plot the basic bar chart we have to follow some steps and they are:
- step1: import matplotlib library
- step2: create some data that you need to plot. The data can be list, arrays, sets, numpy, etc.
- step3: used the `plt.bar(x,y)` function to plot the graph. Here we have to provide what we need on x-axis and y-axis
- step4: here you can add the labels and the title to make the graph more informative using the `plt.xlabel()` and `plt.ylabel()` function and `plt.title()`

```python
#Baic bar graph:
# step1: import the matplotlib library
import matplotlib.pyplot as plt

# step2: lets create some data
subjects = ['Maths','science','history']
marks = [80,70,90]

#step3: plot the bargraph
plt.bar(subjects, marks)
# here to plot the bar chart we use plt.bar(x,y) where the x and y are the values that we need to feed to the system so that it knows what will come on x-axis and what will come on y-axis

# step4: Add labels and title
plt.xlabel('subject')
plt.ylabel('marks')
plt.title('marks of the student')
```

## 1.2 customizing the bar chart:
Here i will use the numpy library to create the data.
- first i will create the data for x - axis using the `np.arange()` function. So it will create number from 1 to 9 on the x-axis.
- for the y-axis i have used the same function but with the step count of 10 that is there will be a difference of 10 in the values of y-axis. [20,30,40,50,60,70,80,90,100]
- Now here i will use the `plt.figure()` function to create the canvas for the plot. It helps control how big the chart should look
- Now i will create an axes object ` ax = plt.axes()`. axes is the area where the graph is actually drawn. The axes helps you customize the plot(background color, grid, ticks, etc)
- now i will set the background color of the graph area as grey using the `ax.set_facecolor()`.
- now i will plot the bar chart using the `plt.bar(x,y)`
- similar to the above code i will set some x and y label.

```python
import numpy as np
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.figure(figsize=(8,5))
ax = plt.axes()
ax.set_facecolor("#ECF0F1")
plt.bar(id1,salary,color="#FFA726")
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.show()
```

## 1.4 Show values on top of each bar
Here to show the values on top of bars, i have used a for loop where i have:
1) used `bar.get_height()`: where it returns the height of the bar.
2) used `plt.text()`: it is used to place text on the graph
3) inside the plt.text(), we have used `bar.get_x()`: this will help in getting the starting x-position of the bar.
4) then we have added bar.get_x() with bar.get_width(): which will give us the width of the bar and to get its center, we have divided it by 2.
5) then we have used yval which helps in placing the text on top of the bar.
6) then to display the text, we have used the `int(yval)` so that the values appear in the form of string.
7) then i have used `va = 'bottom'` here, this will help display the values on top of the bar. there are other two ways which will help display the value, they are 'top' and 'center'
8) Then we come out of the loop and then use the `plt.title()`, `plt.xlabel()`, `plt.ylabel()` to print the values respectively.

```python
barplot = plt.bar(id1,salary)
for bar in barplot:
    yval = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        yval,
        int(yval),
        va='bottom'
    )
plt.title("Bar graph example")
plt.xlabel("Employees")
plt.ylabel("Salary")
```

```python
plt.figure(figsize=(7,7))
barplot = plt.bar(id1,salary)
for bar in barplot:
    yval = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        yval,
        int(yval),
        va = 'bottom'
    )
    plt.yticks([])
plt.title("Bar graph example")
plt.xlabel("Employees")
plt.ylabel("Salary")
```

## 1.8 Stacked Vertical Bar
- It is a type of bar that shows both the total value and the contribution of individual components within each category.
- Here each bar is divided into segments and each segment represents a part of the whole where the full height of the bar is the total value.
- for example we wnat to see the total sales along the year, so here we will have month wise bars where each bar will be divided into online sales, retail sales and wholesale sales.
- This type of bar helps compare total across categories.

```python
x1 = ['akshat','kirtan','jash','rahul']
y1 = [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label="Batting average",width=0.5,color="#ff6f00")
plt.bar(x1,y2,label="bowling average",width=0.5,bottom = y1,color="#ffb300")
for x,y in zip(x1,y1):
    plt.text(x,y-10,'%d'%y,ha='center',va='bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(x,y+z-10,'%d'%y,ha='center',va='bottom')
plt.xlabel("x")
plt.ylabel("y")
plt.title("bar chart")
plt.legend()
plt.show()
```

```python
import numpy as np
x1 = ['akshat','kirtan','jash','rahul']
y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label="Batting average",width=0.5,color="#ff6f00")
plt.bar(x1,y2,label="bowling average",width=0.5,bottom = y1,color="#ffb300")
plt.bar(x1,y3,label="bowling average",width=0.5,bottom = y1+y2,color="#f7dc6f")

for x,y in zip(x1,y1):
    plt.text(x,y-10,'%d'%y,ha='center',va='bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(x,y+z-10,'%d'%y,ha='center',va='bottom')
for x,y,z,p in zip(x1,y3,y1,y2):
    plt.text(x,y+z+p-4,'%d'%y,ha='center',va='bottom')
plt.xlabel("x")
plt.ylabel("y")
plt.title("bar chart")
plt.legend()
plt.show()
```

## 1.9 Grouped Bar Graph
- A grouped bar graph helps when you want to compare multiple related values across the same categories side by side.
- its useful for:
  1) Comparing multiple matrics for the same category.
  2) Easy visual comparison
  3) show trends across categories
  4) better for multiple single bar charts
  5) helps in decision making

```python
import numpy as np
plt.figure(figsize=(7,9))
barwidth = 0.25
y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barwidth for x in pos1]
pos3 = [x + barwidth for x in pos2]

plt.bar(pos1,y1,color="#fbc02d",width=barwidth,label="batting")
plt.bar(pos2,y2,color="#f57f17",width=barwidth,label="bowling")
plt.bar(pos3,y3,color="#e65100",width=barwidth,label="fielding")

plt.xticks(
    [i+barwidth for i in range(len(y1))],
    ['sachin','saurav','rahul','robin']
)
for x,y in zip(pos1,y1):
    plt.text(x,y,'%d'%y,ha='center',va='bottom')
for x,y in zip(pos2,y2):
    plt.text(x,y,'%d'%y,ha='center',va='bottom')
for x,y in zip(pos3,y3):
    plt.text(x,y,'%d'%y,ha='center',va='bottom')
plt.legend()
plt.show()
```

# 3. Scatter Plot
 - Scatter plots is a type of graph used to show the relationship between two continuous numerical variables
 - for x-axis we have 1 numerical variable and for y-axis we have another numerical variable. The dot on the graph represents individual data points.
 - For example x = hours studied and y = exam score. Each dot show how many marks a student scored for a given number of hours.
 - Scatter plot is used to find relationship between variables.
   1) if the graph is positive correlated then we can say that as x increases, y increases.
   2) if the graph is negatively correlated then we can say that as x increases, y decreases.
   3) Not correlated then we can say that the points are scattered randomly.
- Scatter plot helps in detecting the patterns such as linear trends, curved trends or clusters of data
- We can find the outlier in the data using the scatter plot. For example: if a student is studing for 1 hour but scoring 100 marks where as the data shows that maximum students score 40 marks.
- We can compare the large datasets using scatter plot as the data need not be ordered.
- We mainly use scatter plot when:
  1) Both variable are numerical.
  2) We want to check correlation
  3) We are doing feature engineering and EDA
  4) When we want to spot the outliers.

# 4 Pie Chart
- It is a circular chart used to show how a whole is divided into parts. Each slice of the pie represents a category proportion or percentage of the total.
- In pie chart, the entire circle comes up to 100% percent.
- Each slice represents a category share.
- Slice size is proportional to its value.
- Pie chart is used:
  1) To show part to whole relationship.
  2) To compare proportions(not exact value)
- We can use pie chart when data is categorical and we have few categories and values adds up to 100%

```python
import pandas as pd
Goals = [20,12,11,4,3]
Players = ['Ronaldo','Messi','Suarez','Neymar','Salah']
comp = pd.DataFrame({
    "Goals": Goals,
    "players": Players
})
ax = comp.plot(y = "Goals", kind = "pie", labels = comp["players"], autopct = '%1.0f%%', legend = False, title = 'No of goals scored')
ax.set(ylabel="")
```

```python
plt.figure(figsize=(9,9))
area = [48,30,20,15]
labels = ['low','medium','high','critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(area,labels = labels, colors = colors, startangle=45)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(8,8))
area = [48,30,20,15]
total = np.sum(area)
labels = ['low','medium','high','critical']
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x,total*x/100)
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = "true", autopct = val_per)
plt.show()
```

# 6 Histogram
- A hisogram is a graph that shows the distribution of the numerical data.
- Instead of plotting individual data points, it groups data into bins and shows how many data points fall into each range.
- Here the x-axis shows value ranges(bins) and y-axis shows the frequency(count of data product)
- Histogram is used for:
  1) Understanding data distribution(if data is normal, skewed, uniform or has multiple peaks)
  2) It detects patterns
  3) Helps in finding the outliers.

```python
import numpy as np
x = np.random.normal(size=2000)
plt.hist(x,bins=40,color = 'yellowgreen')
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
x = np.random.rand(2000)
plt.hist(x,bins = 30, color = "#D4AC0D")
plt.gca().set(title = "Histogram", ylabel="Frequency")
plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x,bins = 40, color = 'yellowgreen', edgecolor = "#6A9662")
plt.gca().set(title="histogram",ylabel = 'frequency')
plt.show()
```

## 6.1 Binning

```python
plt.figure(figsize = (10,8))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 30, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title='Histogram', ylabel = 'frequency')

plt.figure(figsize = (10,8))
plt.hist(x, bins = 20, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title='Histogram', ylabel = 'frequency')

plt.figure(figsize = (10,8))
plt.hist(x, bins = 10, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title='Histogram', ylabel = 'frequency')
```

# 7 Area Plot
- It is a visualization that shows how quantities change over time and how different categories contribute in total. It is like a line plot +filled color under it.
- Here are plot represents quantitative data where:
  1) x-axis: usually time or sequence.
  2) y-axis: values
  3) Area under the line is filled with color.
  4) often used to show cumulative or stacked value.
- there are 2 types of line graph:
  1) Simple area plot:
     - here its just a line filled with color.
     - Useful when showing trend of one variable.
  2) Stacked area plot:
     Multiple data sets stacked on top of each other which shows trends, how  each categories contribute to the total.
     

```python
import numpy as np
x = np.arange(1,31)
y = np.random.normal(10,11,size=30)
y = np.square(y)
plt.figure(figsize=(16,6))
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
x = np.arange(1,31)
y = np.random.normal(10,11,size=30)
y = np.square(y)
plt.figure(figsize=(16,6))
plt.fill_between(x,y,color="#baf1a1")
plt.plot(x,y,color="#7fcd91")
plt.title("$ Area $ $ chart $",fontsize = 16)
plt.xlabel("$x$",fontsize = 16)
plt.ylabel("$y$",fontsize = 16)
plt.show()
```

## 7.2 Changing fill color and transperancy

```python
x = np.arange(1,31)
y = np.random.normal(10,11,size=30)
y = np.square(y)
plt.figure(figsize=(16,6))
plt.fill_between(x,y,color="#baf1a1",alpha=0.10)
plt.plot(x,y,color="#7fcd91")
plt.title("$ Area $ $ chart $",fontsize = 16)
plt.xlabel("$x$",fontsize = 16)
plt.ylabel("$y$",fontsize = 16)
plt.show()
```

# 8 Contour Plot
- A contour plot is a way to visualize 3-D data on the 2-D graph using countour lines or color-filled regions.
- here x-axis shows horizontal position, y-axis shows vertical position and z-value shows height / intensity / level.
- Its like a topographic map:
  1) Each counter line = same height / same value.
  2) Closely spaced area lines = rapid change.
  3) Wide spacing = slow change.

```python
import numpy as np
def f(x,y):
    return (1-x/2 + x**5 + y**3) * np.exp(-x**2 -y ** 2)
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,alpha = .75, cmap='jet')
C = plt.contour(X,Y,f(X,Y),8,colors='black')
```

# 10 Quiver plot
- A quiver plot shows arrows on a 2D plane. Each arrow represents a vector (direction + magnitude) at a point.
- Every point has (x,y) position and (u,v) vector components.
- It hepls in visualize movement, force, or flow that is things that has both magnitude and direction.

```python
# simple example:
import numpy as np
import matplotlib.pyplot as plt

x = [0,1,2]
y = [0,1,2]

u = [1,1,1]
v = [2,0,-1]

plt.quiver(x,y,u,v)
plt.xlim(-1,3)
plt.ylim(-1,3)

plt.title("Basic quiver plot")
plt.show()
```

```python
n = 8
X,Y = np.mgrid[0:n,0:n]
plt.quiver(X,Y)
plt.show()
```
